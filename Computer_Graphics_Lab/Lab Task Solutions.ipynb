```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
r=100 
ip=[0,r]                        
dp=1-r     
```

```python
xk=ip[0]
yk=ip[1]
p=dp

tx=[]
ty=[]
tx.append(xk)
ty.append(yk)

while((xk<yk)):
    if(p<0):
        p=p+2*xk+3
        xk+=1
        
    else:
        p=p+2*(xk-yk)+5
        xk+=1
        yk-=1
    
    tx.append(xk)
    ty.append(yk)
```

```python
#octant 1
octant=[]
for i in range(0,len(tx)):
    octant.append([tx[i],ty[i]])

#octant 2
t1=tx[::-1]
t2=ty[::-1]
for i in range(0,len(tx)):
    octant.append([t2[i],t1[i]])

#octant 3
t3=np.array(ty)
t4=np.array(tx)*(-1)
for i in range(0,len(tx)):
    octant.append([t3[i],t4[i]])

#octant 4
t5=np.array(tx[::-1])
t6=np.array(ty[::-1])*(-1)
for i in range(0,len(tx)):
    octant.append([t5[i],t6[i]])
    
#octant 5
t7=np.array(tx)*(-1)
t8=np.array(ty)*(-1)
for i in range(0,len(tx)):
    octant.append([t7[i],t8[i]])

#octant 6
t9=np.array(tx[::-1])*(-1)
t10=np.array(ty[::-1])*(-1)
for i in range(0,len(tx)):
    octant.append([t10[i],t9[i]])

#octant 7
t11=np.array(tx)
t12=np.array(ty)*(-1)
for i in range(0,len(tx)):
    octant.append([t12[i],t11[i]])

#octant 8
t13=np.array(tx[::-1])*(-1)
t14=np.array(ty[::-1])
for i in range(0,len(tx)):
    octant.append([t13[i],t14[i]])
```

```python
octant=np.array(octant)
plt.scatter(octant[:,:1], octant[:,1:2], color='b')
plt.plot(octant[:,:1],octant[:,1:2],color='r')
plt.show()
```
