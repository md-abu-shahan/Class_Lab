```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
n=int(input())  #number of test cases
```

```python
for f in range(n):
    print("1st cordinate x, y:")
    x1=float(input())
    y1=float(input())
    print("2nd cordinate x, y:")
    x2=float(input()) 
    y2=float(input()) 
    
    m=(abs(y2-y1))/(abs(x2-x1)) #slope
    
    m2=m
    d1=x1
    d2=y1
    ff=1
    
    if(m2<1):
        siz=(int)(abs(x2-x1))
    else:
        siz=(int)(abs(y2-y1))
    if(x1>x2):
        m=-1*m
        ff=-1
        
    XX=[]
    YY=[]
    print("step-------cordinate-------next cordinate")
    for i in range(0,siz):
        if(m2<1):
            print("| ",i," | ",d1," ",d2," | ",d1+ff," ",d2+m," |");
            d1=d1+ff
            d2=d2+m
            XX.append(d1)
            YY.append(d2)
        else:
            print("| ",i," | ",d1," ",d2," | ",d1+(1/m)," ",d2+ff," |");
            d1=d1+(1/m);
            d2=d2+ff
            XX.append(d1)
            YY.append(d2)
    
    plt.scatter(XX, YY, color='r')
    plt.plot(XX, YY, color='b')
    plt.show()
```
