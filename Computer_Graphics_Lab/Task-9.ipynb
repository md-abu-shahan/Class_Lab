```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
point=[]
P1=[-64,39]
P2=[73,-89]
P3=[-111,-99]
point.append(P1)
point.append(P2)
point.append(P3)
factor=[2,2]
point=np.array(point)
factor=np.array(factor)
```

```python
plt.scatter(point[:,:1],point[:,1:2])
plt.plot(point[:,:1],point[:,1:2])
plt.show()
```

```python
sc=[]
for k in range(0,3):
    sc.append(point[k]*factor)

scaling=np.array(sc)
print(scaling)

plt.scatter(scaling[:,:1],scaling[:,1:2])
plt.plot(scaling[:,:1],scaling[:,1:2])
plt.show()
```
