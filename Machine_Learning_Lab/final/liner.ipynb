```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1.1,1.3,1.5,2.0,2.2,2.3,2.5]
y=[46205,37731, 43525    ,  39891   ,    56642  ,     60150,      54445]
x=np.array(x)
y=np.array(y)
```

```python
m=0
c=0
l=0.0001

run=6000

for i in range(0,run):
    y_cap=[]
    for j in range(0,len(x)):
        y_cap.append(m*x[j]+c)
        
    dm=(-2/(len(x)))*sum(x*(y-y_cap))
    dc=(-2/(len(x)))*sum(y-y_cap)
    
    m=m-l*dm
    c=c-l*dc
print(m,c)

plt.scatter(x, y) 
plt.scatter(x, y_cap, color='red')
plt.plot(x, y_cap, color='red')
plt.show()
```
