```python
g_cmnt=[
     "সজীব স্যার খুব ভালো শিক্ষক",
     "আমি ভালো আছি",
     "আমি নামাজ পড়ি",
     "স্যারের হাসি খুব সুন্দর",
     "আমাদের ডিপার্ট্মেন্ট ভালো"
 ]

b_cmnt=[
     "সে ভাল না",
      "সে খুব খারাপ মানুষ",
     "আমাদের ডিপার্ট্মেন্ট খেলায় খুব খারাপ",
      "আমি খারাপ আছি",
      "রোবট নিজে নিজে কথা বলতে পারে না"
 ]
```

```python
marge_cmnt=g_cmnt.copy()
marge_cmnt.extend(b_cmnt)

g_cmnt_len=len(g_cmnt)
b_cmnt_len=len(b_cmnt)
marge_cmnt_len=len(marge_cmnt)


g_cmnt_count={}
for i in g_cmnt:
    for j in i.split(' '):
        if j in g_cmnt_count:
            g_cmnt_count[j]+=1
        else:
            g_cmnt_count[j]=1
            
            

b_cmnt_count={}
for i in b_cmnt:
    for j in i.split(' '):
        if j in b_cmnt_count:
            b_cmnt_count[j]+=1
        else:
            b_cmnt_count[j]=1
            
            
            
            
marge_cmnt_count={}
for i in marge_cmnt:
    for j in i.split(' '):
        if j in marge_cmnt_count:
            marge_cmnt_count[j]+=1
        else:
            marge_cmnt_count[j]=1
            

g_cmnt_sum=sum(g_cmnt_count.values())
b_cmnt_sum=sum(b_cmnt_count.values())
marge_cmnt_sum=sum(marge_cmnt_count.values())



g_cmnt_p=g_cmnt_len/marge_cmnt_len
b_cmnt_p=b_cmnt_len/marge_cmnt_len


g_cmnt_w_p={}
b_cmnt_w_p={}
marge_cmnt_w_p={}



for i,j in marge_cmnt_count.items():
    marge_cmnt_w_p[i]=(j+1)/(marge_cmnt_sum+2)
    
    
for i,j in b_cmnt_count.items():
    b_cmnt_w_p[i]=(j+1)/(b_cmnt_sum+2)
    

for i,j in g_cmnt_count.items():
    g_cmnt_w_p[i]=(j+1)/(g_cmnt_sum+2)
    
test_case='তারা ভালো বন্ধু'


b_cmnt_u=1
b_cmnt_d=1
for i in test_case.split(' '):
    if i in b_cmnt_w_p:
        b_cmnt_u *= b_cmnt_w_p[i]
        b_cmnt_d *= marge_cmnt_w_p[i]  
    else:
        b_cmnt_u *= (1/(b_cmnt_sum+2))
        b_cmnt_d *= (1/(marge_cmnt_sum+2))
b_cmnt_u *= b_cmnt_p
b_cmnt_ans= b_cmnt_u/b_cmnt_d



g_cmnt_u=1
g_cmnt_d=1
for i in test_case.split(' '):
    if i in g_cmnt_w_p:
        g_cmnt_u *= g_cmnt_w_p[i]
        g_cmnt_d *= marge_cmnt_w_p[i]  
    else:
        g_cmnt_u *= (1/(g_cmnt_sum+2))
        g_cmnt_d *= (1/(marge_cmnt_sum+2))
g_cmnt_u *= g_cmnt_p
g_cmnt_ans= g_cmnt_u/g_cmnt_d


G=(g_cmnt_ans/(g_cmnt_ans+b_cmnt_ans)*100)
B=(b_cmnt_ans/(g_cmnt_ans+b_cmnt_ans)*100)


if(G>B):
    print('Good Comment')
else:
    print('Bed Comment') 
```
