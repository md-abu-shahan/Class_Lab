```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file=pd.read_csv('salary_data.csv')
```

```python
x = file.YearsExperience
y = file.Salary
```

```python
m=0
c=0
l=0.0001

run=60000

for i in range(0,run):
    y_cap=[]
    for j in range(0,len(x)):
        y_cap.append(m*x[j]+c)
        
    dm=(-2/(len(x)))*sum(x*(y-y_cap))
    dc=(-2/(len(x)))*sum(y-y_cap)
    
    m=m-l*dm
    c=c-l*dc
print(m,c)
```

```python
print(y_cap)
```

```python
plt.scatter(x, y) 
plt.scatter(x, y_cap, color='red')
plt.plot(x, y_cap, color='red')
plt.show()
```
